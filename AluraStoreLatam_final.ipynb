### Importación de datos



```python
import pandas as pd

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

tienda = pd.read_csv(url)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)

tienda.head()
```

#1. Análisis de facturación



```python
# Calcular ingresos (sumar columna Precio)
ingreso_tienda1 = tienda["Precio"].sum()
ingreso_tienda2 = tienda2["Precio"].sum()
ingreso_tienda3 = tienda3["Precio"].sum()
ingreso_tienda4 = tienda4["Precio"].sum()

# Resultados
print("Ingreso total Tienda 1:", ingreso_tienda1)
print("Ingreso total Tienda 2:", ingreso_tienda2)
print("Ingreso total Tienda 3:", ingreso_tienda3)
print("Ingreso total Tienda 4:", ingreso_tienda4)
```

# 2. Ventas por categoría

```python
# Agrupar por categoría y contar productos vendidos en cada tienda
ventas_tienda1 = tienda.groupby("Categoría del Producto").size().reset_index(name="Cantidad")
ventas_tienda2 = tienda2.groupby("Categoría del Producto").size().reset_index(name="Cantidad")
ventas_tienda3 = tienda3.groupby("Categoría del Producto").size().reset_index(name="Cantidad")
ventas_tienda4 = tienda4.groupby("Categoría del Producto").size().reset_index(name="Cantidad")

# Mostrar resultados ordenados por popularidad
print("Ventas por categoría - Tienda 1")
print(ventas_tienda1.sort_values(by="Cantidad", ascending=False))

print("\nVentas por categoría - Tienda 2")
print(ventas_tienda2.sort_values(by="Cantidad", ascending=False))

print("\nVentas por categoría - Tienda 3")
print(ventas_tienda3.sort_values(by="Cantidad", ascending=False))

print("\nVentas por categoría - Tienda 4")
print(ventas_tienda4.sort_values(by="Cantidad", ascending=False))
```

# 3. Calificación promedio de la tienda


```python
# Calcular promedio de calificaciones por tienda
promedio_tienda1 = tienda["Calificación"].mean()
promedio_tienda2 = tienda2["Calificación"].mean()
promedio_tienda3 = tienda3["Calificación"].mean()
promedio_tienda4 = tienda4["Calificación"].mean()

print("Calificación promedio - Tienda 1:", round(promedio_tienda1, 2))
print("Calificación promedio - Tienda 2:", round(promedio_tienda2, 2))
print("Calificación promedio - Tienda 3:", round(promedio_tienda3, 2))
print("Calificación promedio - Tienda 4:", round(promedio_tienda4, 2))
```

# 4. Productos más y menos vendidos

```python
import matplotlib.pyplot as plt

def analizar_y_graficar(df, titulo, top_n=10):
    # Agrupar productos y contar ventas
    ventas = df.groupby("Producto").size().reset_index(name="Cantidad")

    # Identificar más y menos vendidos
    mas_vendido = ventas.loc[ventas["Cantidad"].idxmax()]
    menos_vendido = ventas.loc[ventas["Cantidad"].idxmin()]

    # Ordenar para gráfico (Top N)
    ventas_top = ventas.sort_values(by="Cantidad", ascending=False).head(top_n)

    # Crear gráfico horizontal
    plt.figure(figsize=(8,5))
    bars = plt.barh(ventas_top["Producto"], ventas_top["Cantidad"], color="skyblue")
    plt.xlabel("Cantidad Vendida")
    plt.ylabel("Producto")
    plt.title(titulo)
    plt.gca().invert_yaxis()  # Para que el más vendido quede arriba

    # Añadir etiquetas con los valores
    for bar in bars:
        plt.text(bar.get_width() + 0.2, bar.get_y() + bar.get_height()/2,
                 str(int(bar.get_width())), va='center')

    plt.show()

    # Mostrar resultados en consola
    print(f"{titulo}")
    print("Más vendido:", mas_vendido["Producto"], "-", mas_vendido["Cantidad"])
    print("Menos vendido:", menos_vendido["Producto"], "-", menos_vendido["Cantidad"])
    print("-"*50)


# Ejecutar para cada tienda
analizar_y_graficar(tienda, "Tienda 1 - Top 10 Productos")
analizar_y_graficar(tienda2, "Tienda 2 - Top 10 Productos")
analizar_y_graficar(tienda3, "Tienda 3 - Top 10 Productos")
analizar_y_graficar(tienda4, "Tienda 4 - Top 10 Productos")
```

# 5. Envío promedio por tienda




```python
# Calcular promedio de costo de envío por tienda
costo_envio_tienda1 = tienda["Costo de envío"].mean()
costo_envio_tienda2 = tienda2["Costo de envío"].mean()
costo_envio_tienda3 = tienda3["Costo de envío"].mean()
costo_envio_tienda4 = tienda4["Costo de envío"].mean()

print("Costo de envío promedio - Tienda 1:", round(costo_envio_tienda1, 2))
print("Costo de envío promedio - Tienda 2:", round(costo_envio_tienda2, 2))
print("Costo de envío promedio - Tienda 3:", round(costo_envio_tienda3, 2))
print("Costo de envío promedio - Tienda 4:", round(costo_envio_tienda4, 2))
```

# 6. Informe final



# **Introducción**

Durante el desafío, se ayudará al Sr. Juan a decidir qué tienda de su cadena Alura Store debe vender para iniciar un nuevo emprendimiento. Para ello, se analizaron datos de ventas, rendimiento y reseñas de las 4 tiendas de la cadena.

El objetivo es identificar la tienda menos eficiente y presentar una recomendación final, basada en datos objetivos, sobre cuáles tiendas mantener y en cuál dejar de operar.

# **Desarrollo del Análisis**
# **1. Ingresos Totales**
* La Tienda 1 registró los ingresos más altos, superando los 1.150 millones.
* La Tienda 2 se ubicó en segundo lugar, con ingresos sólidos.
* La Tienda 3 presentó ingresos moderados.
* La Tienda 4 obtuvo los menores ingresos, lo que la hace menos competitiva.

```python
# Datos resumidos
tiendas = ["Tienda 1", "Tienda 2", "Tienda 3", "Tienda 4"]
ingresos = [ingreso_tienda1,ingreso_tienda2,ingreso_tienda3, ingreso_tienda4]
calificaciones = [promedio_tienda1,promedio_tienda2,promedio_tienda3,promedio_tienda4]
costos_envio = [costo_envio_tienda1,costo_envio_tienda2,costo_envio_tienda3,costo_envio_tienda4]

colores = ["#8D6E63", "#FF8A65", "#A1887F", "#6D4C41"]

plt.figure(figsize=(8,5))
bars = plt.bar(tiendas, ingresos, color=colores)
plt.title("Ingresos Totales por Tienda", fontsize=14, fontweight="bold")
plt.ylabel("Ingresos", fontsize=12)

# Añadir valores encima de las barras
for bar in bars:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, yval + (yval*0.01),
             f"{yval:,.0f}", ha='center', va='bottom', fontsize=10)

# Quitar bordes superior y derecho
for spine in ["top", "right"]:
    plt.gca().spines[spine].set_visible(False)

plt.show()
```


# **2. Categorías de Productos Más y Menos Vendidas**
* En todas las tiendas predominaron Muebles y Electrónicos, con alta demanda y rotación.
* Las categorías menos vendidas fueron Juguetes y Ropa, que tienen menor impacto en los resultados generales.

# **3. Calificaciones Promedio de los Clientes**
* Todas las tiendas mantuvieron un nivel de satisfacción similar, con calificaciones entre 3.98 y 4.05.
* Tienda 3 obtuvo el promedio más alto (4.05), aunque la diferencia frente a las demás es mínima.
* Tienda 2 y Tienda 4 se mantuvieron prácticamente en el mismo rango (4.04 y 4.00).
* Tienda 1 quedó levemente más baja (3.98), pero sigue estando dentro del rango aceptable.

```python
plt.figure(figsize=(8,5))
bars = plt.bar(tiendas, calificaciones, color=colores)
plt.title("Calificación Promedio de los Clientes", fontsize=14, fontweight="bold")
plt.ylabel("Calificación (1-5)", fontsize=12)
plt.ylim(0, 5)

# Añadir valores encima de las barras
for bar in bars:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, yval + 0.05,
             f"{yval:.2f}", ha='center', va='bottom', fontsize=10)

# Quitar bordes superior y derecho
for spine in ["top", "right"]:
    plt.gca().spines[spine].set_visible(False)

plt.show()
```

# **4. Productos Más y Menos Vendidos**
* En la Tienda 1, los productos más vendidos dominaron el mercado, reflejando gran volumen de compras.
* En las Tiendas 2 y 3, las ventas fueron más equilibradas, mostrando diversificación.
* En la Tienda 4, el volumen de ventas fue bajo en general, lo que limita su atractivo.

# **5. Costo de Envío Promedio**
* La Tienda 1 tuvo el costo de envío más alto, aunque justificado por su gran volumen de ventas.
* La Tienda 2 y la Tienda 3 presentaron costos competitivos y razonables.
* La Tienda 4 tuvo el menor costo de envío, pero esto no compensó sus bajos ingresos y ventas.

```python
plt.figure(figsize=(8,5))
bars = plt.barh(tiendas, costos_envio, color=colores)
plt.title("Costo de Envío Promedio por Tienda", fontsize=14, fontweight="bold")
plt.xlabel("Costo de Envío", fontsize=12)
plt.ylabel("Tiendas", fontsize=12)

# Añadir valores al final de cada barra
for bar in bars:
    xval = bar.get_width()
    plt.text(xval + (xval*0.01), bar.get_y() + bar.get_height()/2,
             f"{xval:,.0f}", ha='left', va='center', fontsize=10)

# Quitar bordes superior y derecho
for spine in ["top", "right"]:
    plt.gca().spines[spine].set_visible(False)

plt.show()
```

# **Conclusión y Recomendación**
Tras integrar todos los factores analizados, se concluye que el Sr. Juan debe concentrar sus operaciones en las Tiendas 1, 2 y 3, y vender la Tienda 4.

* Tienda 1: Lidera en ingresos y volumen de ventas, garantizando rentabilidad.
* Tienda 2: Presenta el mejor equilibrio entre ingresos, satisfacción del cliente y costos de envío.
* Tienda 3: Aunque de menor impacto, ofrece estabilidad y mantiene costos razonables.
* Tienda 4: A pesar de tener el menor costo de envío, sus ingresos y ventas son demasiado bajos, lo que la convierte en la tienda menos eficiente.

En conclusión, se recomienda que el Sr. Juan venda la Tienda 4 y mantenga las Tiendas 1, 2 y 3 como base para el crecimiento futuro de Alura Store.
